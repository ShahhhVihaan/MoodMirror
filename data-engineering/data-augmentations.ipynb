```python
import albumentations as A
import cv2
import os
import numpy as np
import random
```

```python
classes = ["angry","disgust","fear","happy","neutral","sad","surprise"]
```

```python
maxSize = len(os.listdir("happy"))
```

```python
root_output_dir = os.path.join("/content/drive/MyDrive/", "combined-images")
```

```python
transform = A.Compose([
    A.HorizontalFlip(p=0.7),
    A.SafeRotate(p=0.7, limit=30),
    A.RandomScale(p=0.7, scale_limit=(0.0, 0.55)),
    A.ShiftScaleRotate(p=0.7, border_mode=cv2.BORDER_REPLICATE, scale_limit=(0,0), rotate_limit=(0,0), shift_limit=(-0.2, 0.2)),
    A.RandomBrightnessContrast(p=0.7, brightness_limit=(-0.3, 0.3), contrast_limit=(-0.3, 0.3)),
    A.CenterCrop(p=1, width=48, height=48),
])
```

```python
for folder_name in classes:
    files = os.listdir(folder_name)
    num_files = len(files)
    total_new = maxSize - num_files
    indices = np.arange(len(files))
    bucket = set()
    img_counter = 0
    for i in range(total_new):
        if(len(bucket) == num_files):
            bucket.clear()
        file_i = random.choice(indices)
        while(file_i in bucket):
            file_i = random.choice(indices, 1)
        file_name = files[file_i]
        image = cv2.imread(os.path.join(folder_name, file_name))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        transformed = transform(image=image)
        transformed_image = transformed["image"]
        output_dir = os.path.join(root_output_dir, folder_name)
        output_dir = os.path.join(output_dir, folder_name + str(img_counter) + ".jpg")
        cv2.imwrite(output_dir, transformed_image)
        img_counter += 1
    print(folder_name + " files completed")
```
